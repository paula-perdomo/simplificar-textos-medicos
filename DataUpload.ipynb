```python
# Librerias
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos
import boto3
from boto3.s3.transfer import TransferConfig
from google.colab import userdata
import mimetypes
```

### Datos

```python
# Clonar los datos del repositorio
!git clone https://github.com/feliperussi/bridging-the-gap-in-health-literacy
```

```python
AWS_ACCESS_KEY_ID = userdata.get('AWS_ACCESS_KEY_ID_PROJECT')
AWS_SECRET_ACCESS_KEY = userdata.get('AWS_SECRET_ACCESS_KEY_PROJECT')
AWS_DEFAULT_REGION = 'us-east-2'

s3 = boto3.client(
    's3',
    aws_access_key_id=AWS_ACCESS_KEY_ID,
    aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
    region_name=AWS_DEFAULT_REGION
)

# --- Configuración de S3 ---
s3_uri = "s3://maia-grupo-9"
bucket_name = s3_uri.split('/')[2]    # nombre del bucket
s3_prefix = "original_data"           # folder inside S3

# --- Ruta local REAL (del repo clonado) ---
REPO_DIR = "/content/bridging-the-gap-in-health-literacy"
local_folder = os.path.join(REPO_DIR,
    "data_collection_and_processing",
    "Data Sources")                   # ¡ojo! el espacio es parte del nombre

def upload_dir(local_dir, bucket_name, s3_prefix=""):
    if not os.path.isdir(local_dir):
        raise FileNotFoundError(f"Local folder not found: {local_dir}")

    config = TransferConfig(
        multipart_threshold=8 * 1024 * 1024,
        max_concurrency=10,
        multipart_chunksize=8 * 1024 * 1024,
        use_threads=True)

    for root, dirs, files in os.walk(local_dir):
        # Opcional: ignora carpetas típicas
        for skip in [".git", ".ipynb_checkpoints", "__pycache__"]:
            if skip in dirs:
                dirs.remove(skip)

        for f in files:
            local_path = os.path.join(root, f)
            rel_path = os.path.relpath(local_path, start=local_dir)
            key = f"{s3_prefix.rstrip('/')}/{rel_path}".lstrip('/')

            content_type, _ = mimetypes.guess_type(local_path)
            extra_args = {"ContentType": content_type} if content_type else {}

            print(f"Uploading {local_path} -> s3://{bucket_name}/{key}")
            s3.upload_file(local_path, bucket_name, key, ExtraArgs=extra_args, Config=config)

upload_dir(local_folder, bucket_name, s3_prefix)
```

```python
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=s3_prefix)
if 'Contents' in response:
  print('Contents')
  for obj in response['Contents']:
      print(obj['Key'])  # This prints the object key
else:
    print("No objects found in the bucket.")
```
